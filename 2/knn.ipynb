```python
import pandas as pd
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
X = dataset[["Age", "EstimatedSalary"]]
y = dataset[["Purchased"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# without feature scaling

y_pred = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2).fit(X_train, y_train).predict(X_test)
```

```python
# with feature scaling

y_pred = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2).fit(X_train, y_train).predict(scaler.transform(X_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
# without feature scaling

confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# without feature scaling

accuracy_score(y_test, y_pred)
```

```python
# with feature scaling

confusion_matrix(y_test, y_pred)
```

```python
# without feature scaling

accuracy_score(y_test, y_pred)
```
