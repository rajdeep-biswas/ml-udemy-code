Simple Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Salary_Data.csv")
```

```python
X = df[["YearsExperience"]]
y = df[["Salary"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y_pred = LinearRegression().fit(X_train, y_train).predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
plt.scatter(X, y)
plt.plot(X, LinearRegression().fit(X, y).predict(X), 'orange')
plt.show()
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, LinearRegression().fit(X_train, y_train).predict(X_train), 'green')
plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, LinearRegression().fit(X_test, y_test).predict(X_test), 'orange')
plt.show()
```

```python
plt.scatter(X_test, y_pred)
plt.plot(X_test, LinearRegression().fit(X_test, y_test).predict(X_test), 'green')
plt.show()
```

```python
print(LinearRegression().fit(X_train, y_train).coef_)
print(LinearRegression().fit(X_train, y_train).intercept_)
```
